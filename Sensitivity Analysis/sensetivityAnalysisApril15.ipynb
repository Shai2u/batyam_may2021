```python
import pandas as pd
import numpy as np
```

```python
columnsC = ['run','ayearsInBldg','aage','lowDiscount','highDiscount','noDiscount','income','rent','own']
agCols = ['yearsInBldg','age','lowDiscount','highDiscount','noDiscount','income','rent','own']
operations = ['mean','mean','sum','sum','sum','mean','sum','sum']
aggList = dict(zip(agCols,operations))
Sensitivity1 = pd.DataFrame(columns=columnsC)
```

```python
Sensitivity1
```

```python
for i in range(1000):
    Sensitivity1.loc[i,'run'] = i
```

```python
def getAggregated():
    ac.createAgentsAgents()
    return ac.agentsDS.agg(aggList).reset_index()[0].values
```

```python
Sensitivity1 = Sensitivity1.fillna(0)
```

```python
Sensitivity1.loc[0:2].apply(lambda x: getAggregated(),axis=1)
```

```python
#Sensitivity1.iloc[0:10,columnsC[1:]] = Sensitivity1.iloc[0:10].apply(lambda x: getAggregated(),axis=1)
```

```python
for i in Sensitivity1.index:
    ac.createAgentsAgents()
    vals =  ac.agentsDS.agg(aggList).values.tolist()
    Sensitivity1.loc[i,columnsC[1:]] = vals
```
