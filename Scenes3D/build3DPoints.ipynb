```python
import geopandas as gpd
import pyproj
# Pyproj Correction
pyproj.datadir.set_data_dir('/Users/shai/anaconda3/envs/geo_env/share/proj')
import pandas as pd
import numpy as np
from shapely.geometry import Point
```

```python
projRef = pd.read_excel('BuildingInSimulationAndStatsApril12.xlsx')
```

```python
pa = gpd.read_file('PointsAlongApril15.geojson',driver='GeoJSON')
```

```python
pa = pa[pa['project_nu'].isin(projRef['ProjNumber'])].copy()
```

```python
pd.options.display.max_columns =100
pd.options.display.max_rows =100
```

```python
pa.columns
```

```python
pa = pa[['fid', 'Area', 'floors', 'height', 'bld_num', 'bld_addres','hebrew_adr', 'project_nu', 'units', 'project_ty','status','end_date', 'geometry']].copy()
```

```python
pa.reset_index(inplace=True,drop=True)
```

```python
maxFloor = pa['floors'].max().astype(int)+1
minFloor = pa['floors'].min().astype(int)
```

```python
maxFloor
```

```python
for floor in range(minFloor,maxFloor):
    if (floor==1):
        pa.loc[pa['floors']==1,'geometry'] = pa.loc[pa['floors']==1,'geometry'].apply(lambda p: Point(p.x,p.y,6))
    else:
        print('Floor: 1')
        pa.loc[pa['floors']==floor,'geometry'] = pa.loc[pa['floors']==floor,'geometry'].apply(lambda p: Point(p.x,p.y,6))
        subset = pa.loc[pa['floors']==floor].copy()
        for floorItr in range(2,floor+1):
            print('Floor:',floorItr)
            addFloor = subset.copy()
            addFloor['geometry'] = addFloor.apply(lambda p: Point(p['geometry'].x,p['geometry'].y,(p['height']/floor)*floorItr+3),axis=1)
            pa = pa.append(addFloor)
            pa.reset_index(inplace=True,drop=True)
```

```python
pa.to_file('PointAfter3DApril15_1.geojson',driver='GeoJSON')
```
