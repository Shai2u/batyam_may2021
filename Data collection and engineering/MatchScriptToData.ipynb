```python
from_ = ArticleStat.columns.tolist().index('min_living_till_2020')
```

```python
columnsStat = ArticleStat.columns[from_:].tolist()
```

```python
statsRef = ArticleStat[columnsStat].agg(['mean','std']).apply(lambda x: round(x,1))
```

```python
statsRef
```

```python
ArticleMissing.reset_index(inplace=True)
```

```python
ArticleMissing = ArticleMissing.drop(columns={'index'})
```

```python
for col in columnsStat:
    ArticleMissing[col] =ArticleMissing.apply(lambda x: statsRef[col]['mean']+(np.random.rand()-0.5)*2*statsRef[col]['std'],axis = 1)
```

```python
ArticleMissing.iloc[:,2:]
```

```python
ArticleStat.iloc[:,1:].head(2)
```

```python
ArticleStat = pd.concat([ArticleStat,ArticleMissing]).reset_index(drop=True)
```
