## Create Building Mapping Before and After

```python
bldgs_ = gpd.read_file('Json/simulationBldgs_Process_march1_1729.geojson',driver='GeoJSON')
proj_numbers = bldgs_['project_nu'].unique()
bldgMapping = pd.DataFrame(columns=['BeforeBldgs','AfterBldgs','ProjType','ProjNumber'])
c = 0
```

```python
for pNumber in proj_numbers:
    bldgPnumber = bldgs_[bldgs_['project_nu']==pNumber]
    before_ = bldgPnumber[bldgPnumber['status']=='Building before']['bld_addres'].values.tolist()
    after_ = bldgPnumber[bldgPnumber['status']=='Building after']['bld_addres'].values.tolist()
    projType_ = bldgPnumber['project_ty'].iloc[0]
    beforeStr=''
    for item in before_:
        beforeStr = beforeStr+item+','
    afterStr=''
    for item in after_:
        afterStr = afterStr+item+','
    bldgMapping.loc[c,['BeforeBldgs','AfterBldgs','ProjType','ProjNumber']]=[beforeStr[:-1],afterStr[:-1],projType_,pNumber]
    c+=1
```
